```python
# run 94th cell before the execution of current cell
filt = (df['last'] == 'Doe') & (df['first'] == 'John')
```

```python
df.loc[filt, 'email']
```

```python
# run 94th cell before the execution of current cell
filt = (df['last'] == 'Schafer') | (df['first'] == 'John')
```

```python
df.loc[filt, 'email']
```

```python
# run 94th cell before the execution of current cell
filt = (df['last'] == 'Schafer') | (df['first'] == 'John')
```
